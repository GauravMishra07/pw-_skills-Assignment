## Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
df1['Size'].unique()
```

```python
df1['Size']=df1['Size'].str.replace('M','000')
df1['Size']=df1['Size'].str.replace('k','')
df1['Size']=df1['Size'].replace('Varies with device',np.nan)
df1['Size']=df1['Size'].str.replace('+','0')
df1['Size']=df1['Size'].str.replace(',','')

df1['Size'].astype(float)
#df1['Size'].unique()
```

```python
df1["Size"].median()
```

```python
df1["Size"].fillna(df1["Size"].median())
```

```python
plt.figure(figsize=(30,10))
sns.scatterplot(data=df1,x='Size',y='Rating')
```

By above chart we observe that as the size increases rating of an app also increases
